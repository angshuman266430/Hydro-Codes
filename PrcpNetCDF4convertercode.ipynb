```python
import netCDF4
import glob
import xarray as xr
```

```python
for name in glob.glob('C:\data\*.nc4'):
    print(name)
    x_array= xr.open_dataset(name)
    x_array.variables["APCP_surface"][:] = x_array.variables["APCP_surface"][:]*0.5
    if "Aug" in name:
        x_array.to_netcdf(r'C:\data\Aug_Res\\'+name.split("\\")[-1])
    else:
        x_array.to_netcdf(r'C:\data\Sept_Res\\'+name.split("\\")[-1])
    x_array.close()
```
